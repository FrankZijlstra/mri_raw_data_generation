# Interactive demonstration

This runs the interactive visualization provided in the Github repository.

- **Note that because it runs remotely, updates are quite slow (a couple of seconds after each interaction)! Do not drag sliders! Instead click on the point where you want it to end up!**


The visualization provides two sets of 4 sliders at the bottom which can be used to set the latent space coordinates for phase (top set of sliders) and coil sensitivities (bottom set). The display can be switched between showing raw data, coil sensitivity maps, phase maps, and the source magnitude data. It can also be switched between real and generated data. Finally, the latent space coordinates can be fit to the real data with the 'Fit latent' button.

The two sliders at the top allow switching between the top provides scans (index 0 is part of the training set, index 1 is outside the training set), and through the slices of each scan (only 3 slices provided).
